```python
import pandas as pd

income = [478, 512, 196]
expenses = [156, 130, 270]
years = [2018, 2019, 2020]


def create_companyDF(income,expenses,years):
    df = pd.DataFrame({
        'income':income,
        'expenses':expenses
    }, index=years
    )
    return df
df = create_companyDF(income,expenses,years)

def get_profit(df, year):
    if year in df.index:
        result = df.loc[year,'income'] - df.loc[year,'expenses']
        return result
    return None

get_profit(df, 2019)
```
