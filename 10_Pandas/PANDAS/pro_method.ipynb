```python
import pandas as pd

melb_df = pd.read_csv('data/melb_data_fe.csv')

melb_df.head(10)
```

```python
display(melb_df.info())
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'])

melb_df['Quarter'] = melb_df['Date'].dt.quarter

melb_df['Quarter'].value_counts()
```

```python
exclude_list_category = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']

max_uniq = 150

for col in melb_df.columns:
    if melb_df[col].nunique() < max_uniq and col not in exclude_list_category:
        melb_df[col] = melb_df[col].astype('category')
        
melb_df = melb_df.drop(['Quarter'], axis=1)
        
display(melb_df.info())
        
    
    

```
